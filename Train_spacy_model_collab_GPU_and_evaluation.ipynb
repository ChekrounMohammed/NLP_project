```python
import re
def f_remove_accents(old):
    """
    Removes common accent characters, lower form.
    Uses: regex.
    """
    new = old.lower()
    new = re.sub(r'[àáâãäå]', 'a', new)
    new = re.sub(r'[èéêë]', 'e', new)
    new = re.sub(r'[ìíîï]', 'i', new)
    new = re.sub(r'[òóôõö]', 'o', new)
    new = re.sub(r'[ùúûü]', 'u', new)
    return new
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
def TOKENIZE(Text):
    text=word_tokenize(Text)
    return(' '.join(text))
```

```python
text = """ 
— Bakeriella  revoluta  Dub.,  Ann.  Mus.  Col.  Marseille  20  : 28  (1912). 
— Sideroxylon  densiflorum  Bak.  dans  Oliv.,  Fl.  Trop.  Afr.,  3 : 5o3  (1877). 
— Bakerisideroxylon  densiflorum  (Bak.)  Engl.,  le. : 34-
— Vincentella  densiflora  (Bak.)  Pierre,  le. : 37. 
— Vincentella  camerounensis  Pierre  ex Aubr.  et Pellegr.,  Bull.  Soc.  Bot.  Fr. 
"""
text=TOKENIZE(text.lower())
doc = trained_nlp(text)
html = spacy.displacy.render(doc,style="ent")
```
